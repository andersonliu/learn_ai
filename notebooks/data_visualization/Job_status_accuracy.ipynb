```python
import pandas as pd
import matplotlib.pyplot as plt


def plot_line_char(date_file: str):
    # Step 1: Read CSV file into a DataFrame
    df = pd.read_csv(f'data/{date_file}.csv')
    
    # Step 2: Group data by 'type'
    grouped = df.groupby('state')
    
    # Step 3: Plot line chart for each group
    plt.figure(figsize=(10, 6))  # Adjust figure size if needed
    
    for name, group in grouped:
        plt.plot(group['date'], group['count'], label=name)

        # Add value annotations
        for i, row in group.iterrows():
            plt.annotate(row['count'], xy=(row['date'], row['count']), xytext=(5, 5),
                     textcoords='offset points', ha='center', fontsize=8)
    
    plt.xlabel('Date')
    plt.ylabel('Count')
    plt.title(f'Difference on: {date_file}')
    plt.legend()
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.tight_layout()  # Adjust layout
    plt.show()
```

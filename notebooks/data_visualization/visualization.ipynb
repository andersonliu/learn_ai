## Bar Chart

### Load data

```python
# Path of the file to read
flight_filepath = "data/flight_delays.csv"

# Read the file into a variable flight_data
flight_data = pd.read_csv(flight_filepath, index_col="Month")

# Print the data
flight_data
```

### Plot bar chart

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(
    x=flight_data.index,   # the horizontal axis
    y=flight_data['NK']    # the height of each bar
)

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
```

> Note: You must select the indexing column with flight_data.index; flight_data['Month'] will return an error.

## Heatmap

* method: [sns.heatmap()](https://seaborn.pydata.org/generated/seaborn.heatmap.html#seaborn-heatmap)

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(
    data=flight_data, 
    annot=True,        # show number values in each cell
    linewidth=.5,
    fmt=".1f",         # String formatting code to use when adding annotations.
    cmap="YlOrBr",     # Set color palettes. 
)

# Add label for horizontal axis
plt.xlabel("Airline")
```

See avalable [color palettes](https://seaborn.pydata.org/tutorial/color_palettes.html#sequential-color-brewer-palettes) here.

## Scatter Plots

```python
# Path of the file to read
insurance_filepath = "data/insurance.csv"

# Read the file into a variable insurance_data
insurance_data = pd.read_csv(insurance_filepath)

insurance_data.head()
```

Usually, scatter plots are used to highlight the relationship between two continuous variables (like "bmi" and "charges").

We plot scatter chart with method: [sns.scatterplot()](https://seaborn.pydata.org/generated/seaborn.scatterplot.html)

```python
sns.scatterplot(
    data=insurance_data,
    x='bmi', 
    y='charges',
    hue="smoker",   # map smoker value to the color of the points
    style="smoker", # produce points with different markers
)
```

### Plot with a linear regression fit

The scatterplot above suggests that body mass index (BMI) and insurance charges are positively correlated, where customers with higher BMI typically also tend to pay more in insurance costs.

To check the strength of this relationship, we use `sns.regplot()` to add a regression.

```python
sns.regplot(
   data=insurance_data,
    x='bmi', 
    y='charges',
    order=1,        # If order is greater than 1, use a polynomial regression
)
```

### Plot with more linear regression lines

The scatter plot with `hue="smoker"` paramter shows that smokers pays more than non-smokers with increasing BMI.

To further emphasize this fact, we can use the **[sns.lmplot()](https://seaborn.pydata.org/generated/seaborn.lmplot.html)** method to add two regression lines, corresponding to smokers and nonsmokers. (You'll notice that the regression line for smokers has a much steeper slope, relative to the line for nonsmokers!)

```python
sns.lmplot(
    data=insurance_data,
    x='bmi', 
    y='charges',
    order=1,        # If order is greater than 1, use a polynomial regression
    hue="smoker",
)
```

### Categorical scatter plot

Categorical scatter plot used to visualize the relationship between two categorical variables. Unlike numerical scatter plots, which display the relationship between two numerical variables, categorical scatter plots represent the distribution of data points for each category of the variables.

```python
sns.swarmplot(
    data=insurance_data,
    x='smoker',
    y='charges'
)
```

This plot shows us that:

* on average, non-smokers are charged less than smokers, and
* the customers who pay the most are smokers; whereas the customers who pay the least are non-smokers.

## Distributions

```python
# Path of the file to read
iris_filepath = "data/iris.csv"

# Read the file into a variable iris_data
iris_data = pd.read_csv(iris_filepath, index_col="Id")

# Print the first 5 rows of the data
iris_data.head()
```

### Histograms

```python
# Histogram 
sns.histplot(
    data=iris_data, 
    x='Petal Length (cm)'
)
```

### Density plots
`Kernel density estimate (KDE)` plot is a smoothed histogram.

```python
# KDE plot 
sns.kdeplot(
    data=iris_data,
    x='Petal Length (cm)', 
    fill=True,              # fill a color in the area below the curve
)
```

In fact, `sns.histplot()` can plat the kde too. 
Add a kernel density estimate to smooth the histogram, providing complementary information about the shape of the distribution:

```python
# Histogram 
sns.histplot(
    data=iris_data, 
    x='Petal Length (cm)',
    kde=True,
)
```

### 2D KDE plots

We can create a two-dimensional (2D) KDE plot with the `sns.jointplot()` command.

In the plot below, the color-coding shows us how likely we are to see different combinations of sepal width and petal length, where darker parts of the figure are more likely.

```python
# 2D KDE plot
sns.jointplot(
    data=iris_data,
    x='Petal Length (cm)', 
    y='Sepal Width (cm)', 
    kind="kde",
    fill=True, thresh=0, cmap='Blues' # plot with color mapping
)
```

In addition to the 2D KDE plot in the center, the curves at the top and right of the figure are KDE plots too:

* The top KDE plot for the data on the x-axis (`Petal Length (cm)`), and
* The right KDE plot for the data on the y-axis (`Sepal Width (cm)`).

### Color-coded plots

We will create plots to understand differences between the species.

By adding `hue='Species'`, the histogram color coded on species.

```python
# Histogram 
sns.histplot(
    data=iris_data, 
    x='Petal Length (cm)',
    kde=True,
    hue='Species',  # color colded on Species categorial
)
```

We can also create a KDE plot for each species by using `sns.kdeplot` (as above). 

```python
# KDE plot 
sns.kdeplot(
    data=iris_data,
    x='Petal Length (cm)', 
    fill=True,              # fill a color in the area below the curve
    hue='Species',          # color colded on Species categorial
)
```

See the diagram above, we might be able to classify any iris plant as Iris setosa just by looking at the petal length: 
* If the petal length of an iris flower is less than 2 cm, it's most likely to be Iris setosa!

## Summary of Plot Types
![plot_types](image/plot_types.png)

we've broken the chart types into three broad categories to help with this.

* **Trend**s - A trend is defined as a pattern of change.
  * `sns.lineplot` - **Line charts** are best to show trends over a period of time, and multiple lines can be used to show trends in more than one group.

* **Relationship** - There are many different chart types that you can use to understand relationships between variables in your data.
  * `sns.barplot` - **Bar charts** are useful for comparing quantities corresponding to different groups.
  * `sns.heatmap` - **Heatmaps** can be used to find color-coded patterns in tables of numbers.
  * `sns.scatterplot` - **Scatter plots** show the relationship between two continuous variables; if color-coded, we can also show the relationship with a third [categorical variable](https://en.wikipedia.org/wiki/Categorical_variable).
  * `sns.regplot` - Including a **regression line** in the scatter plot makes it easier to see any linear relationship between two variables.
  * `sns.lmplot` - This command is useful for drawing multiple regression lines, if the scatter plot contains multiple, color-coded groups.
  * `sns.swarmplot` - **Categorical scatter plots** show the relationship between a continuous variable and a categorical variable.

* **Distribution** - We visualize distributions to show the possible values that we can expect to see in a variable, along with how likely they are.
  * `sns.histplot` - **Histograms** show the distribution of a single numerical variable.
  * `sns.kdeplot` - **KDE plots** (or **2D KDE plots**) show an estimated, smooth distribution of a single numerical variable (or two numerical variables).
  * `sns.jointplot` - This command is useful for simultaneously displaying a 2D KDE plot with the corresponding KDE plots for each individual variable.
